```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
name = '2010CO1'
dat1 = np.loadtxt(f'{name}dat.txt')
dato = np.loadtxt(f'newepoch_{name.lower()}.txt')
ll = len(dat1)
dat0 = dato[:len(dato)//2]
a1,e1,del1,r1,alp1 = dat1[:,0:3],dat1[:,3:6],dat1[:,6],dat1[:,7],dat1[:,8]
a0,e0 = dat0[:,0:3],dat0[:,3:6]
del0 = np.sqrt((a0[:,0]-e0[:,0])**2+(a0[:,1]-e0[:,1])**2+(a0[:,2]-e0[:,2])**2)
r0 = np.sqrt(e0[:,0]**2+e0[:,1]**2+e0[:,2]**2)
ra0 = np.sqrt(a0[:,0]**2+a0[:,1]**2+a0[:,2]**2)
alp0 = np.arccos((del0**2+ra0**2-r0**2)/(2*del0*ra0))
alp0 = 180*alp0/3.14
```

```python
plt.subplot(2,1,1)
plt.plot(del1[1:ll-1],label='dat')
plt.plot(del0[1:ll-1],label = 'jpl')
plt.xlabel('n')
plt.ylabel('au')
plt.title('Delta (au)')
plt.legend()
plt.subplot(2,1,2)
plt.plot(del1[1:ll-1]-del0[1:ll-1],label='dat-jpl')
plt.xlabel('n')
plt.ylabel('au')
plt.legend()
plt.savefig(f'{name}_delta.png')
```

```python
plt.subplot(2,1,1)
plt.plot(r1[1:ll-1],label='dat')
plt.plot(ra0[1:ll-1],label = 'jpl')
plt.xlabel('n')
plt.ylabel('au')
plt.title('rhelio(au)')
plt.legend()
plt.subplot(2,1,2)
plt.plot(r1[1:ll-1]-ra0[1:ll-1],label='dat-jpl')
plt.xlabel('n')
plt.ylabel('au')
plt.legend()
plt.savefig(f'{name}_helio.png')
```

```python
plt.subplot(2,1,1)
plt.plot(alp1[1:ll-1],label='dat')
plt.plot(alp0[1:ll-1],label = 'jpl')
plt.xlabel('n')
plt.ylabel('au')
plt.title('alpha(deg)')
plt.legend()
plt.subplot(2,1,2)
plt.plot(alp1[1:ll-1]-alp0[1:ll-1],label='alpha:dat-jpl')
plt.xlabel('n')
plt.ylabel('au')
plt.legend()
plt.savefig(f'{name}_alpha.png')
```
