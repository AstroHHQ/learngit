## 对导出的txt处理

```python
import numpy as np
import pandas as pd
from urllib.request import urlopen
from requests.api import request
from tqdm import tqdm
data = pd.read_table('wise.txt')
```

```python
data
```

将其中字符串转换为数字

```python
df = data
#a = int(df["WiseData_time"][4][0:4])
#df["WiseData_time"][4] = a
#df
for i in tqdm(range(len(df))):
    dfi = df["WiseData_time"][i]
    if isinstance(dfi,str):
        numdf = int(dfi[0:4])
        df["WiseData_time"][i] = numdf
df
```

对数据进行排序

```python
df2 = df.sort_values(by='WiseData_time',ascending=False)
df2.to_csv('wise0.txt',sep='\t',index=False)
```
