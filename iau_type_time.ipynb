## 对导出的txt处理  
输出了最终结果，含wise数据以及type+是否是pha

```python
import numpy as np
import pandas as pd
from urllib.request import urlopen
from requests.api import request
from tqdm import tqdm
data = pd.read_table('wise.txt')
```

```python
data
```

将其中字符串转换为数字

```python
df = data
#a = int(df["WiseData_time"][4][0:4])
#df["WiseData_time"][4] = a
#df
for i in tqdm(range(len(df))):
    dfi = df["WiseData_time"][i]
    if isinstance(dfi,str):
        numdf = int(dfi[0:4])
        df["WiseData_time"][i] = numdf
df
```

合并统计的类型以及是否是pha

```python
data_raw = pd.read_table('mpc_wise_type_nha.txt')
data_raw
```

```python
data_final = data_raw.loc[:,['Name','Diam','type','PHA']]
data_final['time'] = df['WiseData_time']
data_sort =data_final.sort_values(by='time',ascending=False)
data_final.to_csv('mpc_final.txt',sep='\t',index=False)
data_sort.to_csv('mpc_final_sort.txt',sep='\t',index=False)
```
