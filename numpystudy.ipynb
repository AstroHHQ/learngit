```python
#类似excel的seris
dic1 = {}
dic2 = {}
dic3 = {}
dic4 = {}
dic5= {}
np.random.seed(0)
rdi = np.random.randint(10,100,(10))

for li in rdi:
    dic1[li] = li//10
    dic2[li] = li%10
    dic3[li] = li*2
    if li%2==0:
        dic4[li] = 1
    else:
        dic5[li] = 1
#print('dic1:',dic1,'rdi',rdi)
sr10 = pd.Series(dic1)
sr1 = pd.Series(dic2)
sr3 = pd.Series(dic3)
sr4 = pd.Series(dic4)
sr5 = pd.Series(dic5)
#print('sr10:\n',sr1)
df1 = pd.DataFrame({'十':sr10,'个':sr1,2:sr3,22:sr4,33:sr5})
df1['个']
df1['十']
df1.iloc[:3,:1]
df1.loc[:77,:33]
```
