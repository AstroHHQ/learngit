```python
import numpy as np
import pandas as pd
from tqdm import tqdm
from astropy.time import Time 
```

```python
astname = '2010fh81'
fname = f'time_{astname}.txt'
rawdata = pd.read_csv(fname)
len(rawdata)
```

```python
time_tdb = []
time_utc = []
for i in range(len(rawdata)):
    ti = Time(rawdata['mjd'][i],format='mjd',scale='utc')
    time_tdb.append(ti.tdb.jd)
    time_utc.append(ti.utc.jd)
```

```python
tt = np.array(time_tdb[:15])
np.savetxt(f'{astname}_tdb_time.txt',tt,fmt='%.6f')
tt2 = np.array(time_utc[:15])
np.savetxt(f'{astname}_utc_time.txt',tt2,fmt='%.6f')
```
